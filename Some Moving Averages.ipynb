```python
import numpy as np 
import pandas as pd 
import yfinance as yf
import matplotlib.pyplot as plt

# ^GSPC is the symbol of S&P 500 in Yahoo! Finance
tickerData = yf.Ticker('^GSPC')

df = tickerData.history(period='1d', start='2010-1-1', end='2015-10-1')

df.head()

close = df['Close']

all_weekdays = pd.date_range(start='2010-1-1', end='2015-10-1', freq='B')

close = close.reindex(all_weekdays)

close = close.fillna(method='ffill')

print(all_weekdays)

close.describe()

plt.plot(close)
```

```python
df['Weekavg'] = df.Close.rolling(7).mean()
df['Monthavg'] = df.Close.rolling(30).mean()
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(df['Close'], label='S&P 500', color='k')
plt.plot(df['Weekavg'], label='7 day average', color='b')
plt.plot(df['Monthavg'], label='30 day average', color='m')
plt.legend(loc=2)
```
