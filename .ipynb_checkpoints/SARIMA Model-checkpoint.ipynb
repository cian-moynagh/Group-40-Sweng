```python
import numpy as np 
import pandas as pd 
import yfinance as yf
import matplotlib.pyplot as plt
import statsmodels.api as stat#

def decompose(data):
    res = stat.tsa.seasonal_decompose(data.Close.dropna(),period=365)
    fig = res.plot()
    fig.set_figheight(8)
    fig.set_figwidth(15)
    plt.show()
    
    
    #train_test_split
    tr_start,tr_end = '2011-01-01','2014-09-30'
    te_start,te_end = '2014-10-01','2014-12-31'
    tra = data['Close'][tr_start:tr_end].dropna()
    tes = data['Close'][te_start:te_end].dropna()
    
    #ADF-test(Original-time-series)
    res = stat.tsa.adfuller(data['Close'].dropna(),regression='ct')
    print('p-value:{}'.format(res[1]))
    
    #ADF-test(differenced-time-series)
    res = stat.tsa.adfuller(data['Close'].diff().dropna(),regression='c')
    print('p-value:{}'.format(res[1]))
    
    
    
    #we use tra.diff()(differenced data), because this time series is unit root process.
    fig,ax = plt.subplots(2,1,figsize=(20,10))
    fig = stat.graphics.tsa.plot_acf(tra.diff().dropna(), lags=50, ax=ax[0])
    fig = stat.graphics.tsa.plot_pacf(tra.diff().dropna(), lags=50, ax=ax[1])
    plt.show()
    
    
    

def sarima_model(data):
    
    #train_test_split
    tr_start,tr_end = '2011-01-01','2014-09-30'
    te_start,te_end = '2014-10-01','2014-12-30'
    tra = data['Close'][tr_start:tr_end].dropna()
    tes = data['Close'][te_start:te_end].dropna()
    
    the_order = (1, 1, 1)
    the_seasonal_order = (1, 1, 1, 12)
    
    model = stat.tsa.statespace.SARIMAX(tra,order=the_order,seasonal_order=the_seasonal_order)
                                #enforce_stationarity=False, enforce_invertibility=False,freq='D').fit()
    result = model.fit()
    print(result.summary())
    
    res = result.resid
    fig,ax = plt.subplots(2,1,figsize=(15,8))
    fig = stat.graphics.tsa.plot_acf(res, lags=50, ax=ax[0])
    fig = stat.graphics.tsa.plot_pacf(res, lags=50, ax=ax[1])
    plt.show()
    
    forecasting = result.forecast()
    
    from sklearn.metrics import mean_squared_error
    pred = result.predict(tr_end,te_end)[1:]
    print('SARIMA model MSE:{}'.format(mean_squared_error(tes,pred)))
    
    
```

```python
# ^GSPC is the symbol of S&P 500 in Yahoo! Finance
tickerData = yf.Ticker('^GSPC')

df = tickerData.history(period='1d', start='2010-1-1', end='2015-10-1')

df.head()

plt.plot(df.Close)

decompose(df)

sarima_model(df)
```
